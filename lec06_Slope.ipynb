## slope, დახრილობა

Ფიზიკურ სამყაროში ერთ-ერთი ყველაზე ფუნდამენტური და მნიშვნელოვანი საკვლევი ასპექტი არის **ცვლილება** - რატომ, რაზე დაყრდნობით, რის მიხედვით, რამდენად და ა.შ. იცვლება ესა თუ ის გარემოება.

Წინა ლექციაში ჩვენ ვახსენეთ slope, დახრილობა, როგორც იმის საზომი, თუ რამდენად იცვლება y x-ის ცვლილებასთან ფარდობითად . ამ ლექციაში დეტალურად გავაანალიზებთ დახრილობის მნიშვნელობას.


image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
%matplotlib inline

def plot_curve_and_tangent(x_point):
    def f(x):
        return x**2

    def tangent_line(x, x0):
        return 2*x0*(x - x0) + f(x0)

    x = np.linspace(0, 3, 1000)
    y = f(x)

    fig, ax = plt.subplots(figsize=(10, 8))

    # მრუდი
    ax.plot(x, y, 'k', linewidth=2)

    # მხები წერტილი
    ax.plot(x_point, f(x_point), 'o', color='orange', markersize=10)

    # მხები ხაზი
    x_tangent = np.linspace(0, x_point*2, 100)
    y_tangent = tangent_line(x_tangent, x_point)
    ax.plot(x_tangent, y_tangent, color='orange', linewidth=2)

    # სამკუთხედი
    ax.fill_between([0, x_point, x_point], [0, 0, f(x_point)], color='blue', alpha=0.3)

    # ვერტიკალური და ჰორიზონტალური ხაზები
    ax.plot([x_point, x_point], [0, f(x_point)], 'b--', linewidth=1)
    ax.plot([0, x_point], [f(x_point), f(x_point)], 'b--', linewidth=1)

    ax.set_xlim(0, 3)
    ax.set_ylim(0, 9)
    ax.set_aspect('equal')
    ax.axis('off')

    # წითელი ხაზი ქვემოთ
    ax.plot([0, 3], [-0.5, -0.5], color='red', linewidth=10)

    plt.tight_layout()
    plt.show()

interactive_plot = interactive(plot_curve_and_tangent,
                               x_point=FloatSlider(min=0.1, max=3, step=0.1, value=1.5,
                                                   description='x point:'))
interactive_plot
```

## Მაგრამ, რა უნდა გავაკეთოთ, თუ გვაინტერესებს slope **კონკრეტულ** წერტილში ? მაგ. (1;1) - ში

image.png

Ამისთვის უნდა გავავლოთ ხაზი ზუსტად ამ წერტილზე.
Მაგრაამ
Შეგვიძლია თუ არა იმავენაირად გავზომოთ დახრილობა?
ჩვენი ფორმულით 1;1 წერტილისთვის გამოგვივა 0
 დახრილობა ნამდვილად რაღაც დადებითი რიცხვი უნდა იყოს (ვიზუალურად ვხედავთ) ,

სინამდვილეში, რაღაც სხვა გვჭირდება კონკრეტულ წერტილში slope-ის გამოსაანგარიშებლად.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
%matplotlib inline

def plot_derivative(x1, x2):
    x = np.linspace(0, 3, 300)
    y = x**2  # ფუნქცია A(x) = x^2

    y1, y2 = x1**2, x2**2

    fig, ax = plt.subplots(figsize=(10, 8))

    # ფუნქციის გრაფიკი
    ax.plot(x, y, 'b-', label='A(x) = x²')

    # წერტილები და მათი შემაერთებელი წრფე
    ax.plot([x1, x2], [y1, y2], 'y-', linewidth=2)
    ax.plot(x1, y1, 'bo', markersize=8)
    ax.plot(x2, y2, 'bo', markersize=8)

    # Δx და ΔA-ს ვიზუალიზაცია
    ax.plot([x1, x1], [y1, y2], 'k--')
    ax.plot([x1, x2], [y1, y1], 'k--')

    # ანოტაციები
    ax.annotate(f'Δx = {x2-x1:.2f}', xy=((x1+x2)/2, y1), xytext=((x1+x2)/2, y1-0.5),
                ha='center', va='top', arrowprops=dict(arrowstyle='->'))
    ax.annotate(f'ΔA = {y2-y1:.2f}', xy=(x1, (y1+y2)/2), xytext=(x1-0.3, (y1+y2)/2),
                ha='right', va='center', arrowprops=dict(arrowstyle='->'))

    slope = (y2 - y1) / (x2 - x1)
    ax.set_title(f'წარმოებულის ვიზუალიზაცია\nდახრილობა = ΔA / Δx = {slope:.2f}')
    ax.set_xlabel('x')
    ax.set_ylabel('A(x)')
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 9)
    ax.grid(True)

    plt.tight_layout()
    plt.show()

    print(f"საშუალო დახრილობა: {slope:.2f}")
    print(f"x₁ წერტილში წარმოებული: {2*x1:.2f}")
    print(f"x₂ წერტილში წარმოებული: {2*x2:.2f}")

interactive_plot = interactive(plot_derivative,
                               x1=FloatSlider(min=0.1, max=2.5, step=0.1, value=1),
                               x2=FloatSlider(min=0.2, max=2.9, step=0.1, value=2))
display(interactive_plot)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive, FloatSlider, VBox, HBox
%matplotlib inline

def plot_slope_and_derivative(x1, delta_x):
    x2 = x1 + delta_x
    y1 = x1**2
    y2 = x2**2
    delta_y = y2 - y1
    slope = delta_y / delta_x if delta_x != 0 else 0

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    # გრაფიკი 1: ფუნქცია და დახრილობა
    x = np.linspace(0, 10, 1000)
    y = x**2

    ax1.plot(x, y, 'b-', label='y = x²')
    ax1.plot([x1, x2], [y1, y2], 'r-', linewidth=2)
    ax1.plot([x1, x1], [y1, y2], 'g--')
    ax1.plot([x1, x2], [y1, y1], 'g--')

    ax1.scatter([x1, x2], [y1, y2], color='red', s=50)
    ax1.annotate(f'({x1:.2f}, {y1:.2f})', (x1, y1), xytext=(5, 5), textcoords='offset points')
    ax1.annotate(f'({x2:.2f}, {y2:.2f})', (x2, y2), xytext=(5, 5), textcoords='offset points')

    ax1.set_xlim(0, 10)
    ax1.set_ylim(0, 100)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title('ფუნქცია და დახრილობა')
    ax1.legend()

    # გრაფიკი 2: კვადრატის ფართობის ანალოგია
    ax2.add_patch(plt.Rectangle((0, 0), x1, x1, facecolor='blue', alpha=0.5))
    ax2.add_patch(plt.Rectangle((x1, 0), delta_x, x1, facecolor='red', alpha=0.5))
    ax2.add_patch(plt.Rectangle((0, x1), x1, delta_x, facecolor='red', alpha=0.5))
    ax2.add_patch(plt.Rectangle((x1, x1), delta_x, delta_x, facecolor='yellow', alpha=0.5))

    ax2.text(x1/2, x1/2, f'x²\n{y1:.2f}', ha='center', va='center')
    ax2.text(x1 + delta_x/2, x1/2, f'Δx·x\n{delta_x*x1:.2f}', ha='center', va='center')
    ax2.text(x1/2, x1 + delta_x/2, f'Δx·x\n{delta_x*x1:.2f}', ha='center', va='center')
    ax2.text(x1 + delta_x/2, x1 + delta_x/2, f'(Δx)²\n{delta_x**2:.2f}', ha='center', va='center')

    ax2.set_xlim(0, 10)
    ax2.set_ylim(0, 10)
    ax2.set_aspect('equal')
    ax2.set_title('კვადრატის ფართობის ანალოგია')

    plt.tight_layout()
    plt.show()

    print(f"x1 = {x1:.2f}, x2 = {x2:.2f}")
    print(f"y1 = {y1:.2f}, y2 = {y2:.2f}")
    print(f"Δx = {delta_x:.2f}")
    print(f"Δy = {delta_y:.2f}")
    print(f"დახრილობა (Δy / Δx) = {slope:.2f}")
    print(f"წარმოებული x = {x1:.2f} წერტილში: {2*x1:.2f}")
    print(f"ფართობის ცვლილება: {delta_y:.2f}")
    print(f"მიახლოებითი ცვლილება (2x·Δx): {2*x1*delta_x:.2f}")

interactive_plot = interactive(plot_slope_and_derivative,
                               x1=FloatSlider(min=0.1, max=9, step=0.1, value=2, description='x:'),
                               delta_x=FloatSlider(min=0.1, max=2, step=0.1, value=0.5, description='Δx:'))

display(VBox([interactive_plot]))
```
